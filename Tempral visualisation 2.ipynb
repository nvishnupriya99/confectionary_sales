```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib inline
```

```python
GM_stock = pd.read_csv('GM.csv',
                         index_col='Date',
                         parse_dates=True)
GM_stock.head(10)
```

```python
tsla_stock = pd.read_csv('TSLA.csv',
                         index_col='Date',
                         parse_dates=True)
tsla_stock.head(10)
```

```python
bmw_stock = pd.read_csv('BMW.csv',
                         index_col='Date',
                         parse_dates=True)
bmw_stock.head(10)
```

```python
ford_stock = pd.read_csv('Ford.csv',
                         index_col='Date',
                         parse_dates=True)
ford_stock.head(10)
```

```python
plt.figure(figsize=(12, 8), dpi=150)

tsla_stock['Open'].plot(label='tsla_stock', color='red')
bmw_stock['Open'].plot(label='BMW_stock')
GM_stock['Open'].plot(label='GM_shares')

plt.title('Price Plot')

plt.xlabel('Years')
plt.legend()
```

```python
# Aligning index
ford_stock["Close"] = tsla_stock.Close

# removing Missing Values
ford_stock.dropna(inplace=True)

ford_stock.head(5)
```

```python
plt.figure(figsize=(12, 8), dpi=150)

tsla_stock['Open'].plot(label='tsla_stock', color='red')
ford_stock['Open'].plot(label='Ford_stock')

plt.title('Price Plot')

plt.xlabel('Years')
plt.legend()
```
