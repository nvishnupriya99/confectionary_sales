```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
# Load the breast cancer dataset
cancer_data = load_breast_cancer()

# Convert the dataset into a DataFrame for easier manipulation
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df['target'] = cancer_data.target
```

```python
df.head()
```

```python
# Set up seaborn style
sns.set(style="whitegrid")
```

```python
# Plot 1: Distribution of a single feature (e.g., 'mean radius')
plt.figure(figsize=(10, 6))
sns.histplot(df['mean radius'], kde=True, color='blue')
plt.title('Distribution of Mean Radius')
plt.xlabel('Mean Radius')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Boxplot of a single feature (e.g., 'mean texture')
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='mean texture', data=df, palette='Set2')
plt.title('Boxplot of Mean Texture by Cancer Type')
plt.xlabel('Cancer Type (0 = Malignant, 1 = Benign)')
plt.ylabel('Mean Texture')
plt.show()
```

```python
# Plot 3: Violin plot for a feature (e.g., 'mean smoothness') based on the target
plt.figure(figsize=(10, 6))
sns.violinplot(x='target', y='mean smoothness', data=df, palette='Set3')
plt.title('Violin Plot of Mean Smoothness by Cancer Type')
plt.xlabel('Cancer Type (0 = Malignant, 1 = Benign)')
plt.ylabel('Mean Smoothness')
plt.show()
```

```python
# Plot 4: Strip plot for a feature (e.g., 'mean compactness') based on the target
plt.figure(figsize=(10, 6))
sns.stripplot(x='target', y='mean compactness', data=df, jitter=True, palette='Set1')
plt.title('Strip Plot of Mean Compactness by Cancer Type')
plt.xlabel('Cancer Type (0 = Malignant, 1 = Benign)')
plt.ylabel('Mean Compactness')
plt.show()
```

```python
# Plot 5: Swarm plot of 'mean radius' by target
plt.figure(figsize=(10, 6))
sns.swarmplot(x='target', y='mean radius', data=df, palette='Set2')
plt.title('Swarm Plot of Mean Radius by Cancer Type')
plt.xlabel('Cancer Type (0 = Malignant, 1 = Benign)')
plt.ylabel('Mean Radius')
plt.show()
```
