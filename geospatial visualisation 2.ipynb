```python
import seaborn as sns

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('US_elect_county.csv')
df.head()
```

```python
fig = px.choropleth (df, 
                     locations = 'State Postal', 
                     locationmode= 'USA-states',
                     scope= 'usa',
                     color= 'percent_republican',
                     hover_name= 'County Name',
                     hover_data= ['Obama vote', 'Romney vote', 'percent_democrat', 'percent_republican'],
                     range_color=[10, 90],
                     color_continuous_scale = 'viridis',
                     title= 'US 2012 Elections Results')
fig.show()
```
