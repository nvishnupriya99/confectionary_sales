```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tsla_df = pd.read_csv ('TSLA.csv', index_col= 0, parse_dates= True)
tsla_df = tsla_df. reset_index ()
tsla_df. head()
```

```python
tsla_df_small = tsla_df [-30:]
tsla_df_small
```

```python
green_df = tsla_df_small [tsla_df_small.Close > tsla_df_small. Open].copy()
green_df ['Height'] = green_df ['Close'] - green_df ['Open']. copy()

red_df = tsla_df_small [tsla_df_small. Close < tsla_df_small. Open].copy()
red_df ['Height'] = red_df ['Close'] - red_df ['Open']. copy()
```

```python
green_df. head()
```

```python
# Set up the figure with specified size
fig = plt.figure(figsize=(15, 7))

# Vertical lines for Tesla data from 'Low' to 'High' prices
plt.vlines(x=tsla_df_small['Date'], ymin=tsla_df_small['Low'], ymax=tsla_df_small['High'],
           color='grey')

# Green bars for 'Open' to 'Height' values
plt.bar(x=green_df['Date'], height=green_df['Height'], bottom=green_df['Open'],
        color='green')

# Red bars for 'Close' to 'Height' values
plt.bar(x=red_df['Date'], height=red_df['Height'], bottom=red_df['Close'], color='red')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price (Â£)')
plt.title('Tesla Data')

# Display the plot
plt.show()
```
