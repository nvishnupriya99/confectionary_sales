```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import geodatasets
import folium

plt. rcParams ['figure.dpi'] = 100
```

```python
countries = gpd.read_file("110m_cultural/ne_110m_admin_0_countries.shp")

# Display the first few rows
print(countries.head())
```

```python
countries.loc [countries ['SOVEREIGNT']== 'Canada'].plot()
```

```python
countries.loc [countries ['SOVEREIGNT']== 'United States of America'].plot()
```

```python
# Load the cities data
cities = gpd.read_file("110m_cultural/ne_110m_populated_places_simple.shp")
cities.head()
```

```python
cities.plot()
```

```python
ax = countries.plot(figsize = (10, 10))
cities.plot (ax = ax, color = 'cyan', markersize = 15)
```

```python
ax = countries.plot(figsize = (15, 15))
cities.plot (ax = ax, color = 'cyan', markersize = 10)
```

```python
# Plot the countries
fig, ax = plt.subplots(figsize=(15, 15))
countries.boundary.plot(ax=ax, linewidth=1)

# Plot the cities with city sizes proportional to pop_max
cities.plot(ax=ax, color='yellow', markersize=cities['pop_max'] / 100000 , legend=True)

# Add title and show the plot
plt.title("Countries and Cities with Population Size")
plt.show()
```

```python
# Plot the countries, coloring them by population
fig, ax = plt.subplots(figsize=(15, 15))
countries.plot( ax=ax, cmap='OrRd', linewidth=0.8, edgecolor='0.8', legend=True)

# Plot the cities with sizes proportional to 'pop_max'
cities.plot(ax=ax, color='yellow', markersize=cities['pop_max'] / 100000)

# Add title and show the plot
plt.title("Countries and Cities with Population Size")
plt.show()
```
