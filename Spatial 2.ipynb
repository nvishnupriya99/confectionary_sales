```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
# Load the breast cancer dataset
cancer_data = load_breast_cancer()

# Convert the dataset into a DataFrame
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df['target'] = cancer_data.target

# Set up seaborn style
sns.set(style="whitegrid")
```

```python
df.head()
```

```python
# Plot 1: Scatter plot of 'mean radius' vs. 'mean texture'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mean radius', y='mean texture', hue='target', palette='coolwarm', data=df)
plt.title('Scatter Plot of Mean Radius vs. Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.legend(title='Cancer Type', labels=['Malignant', 'Benign'])
plt.show()
```

```python
# Plot 2: Pair plot of selected features
selected_features = ['mean radius', 'mean texture', 'mean smoothness', 'mean compactness']
sns.pairplot(df[selected_features + ['target']], hue='target', palette='coolwarm', diag_kind='kde')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
# Plot 3: Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
# Plot: 2D Histogram of 'mean radius' vs. 'mean texture'
plt.figure(figsize=(10, 8))
plt.hist2d(df['mean radius'], df['mean texture'], bins=30, cmap='viridis')
plt.colorbar(label='Counts')
plt.title('2D Histogram of Mean Radius vs. Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.show()
```
