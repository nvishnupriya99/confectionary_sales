```python
#Load the seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
#check on shaep of data, has five features and 150 observations.
iris.shape
```

```python
#Checking first few rows/tuples of data
iris.head()
```

Proximity:	
Explain how objects that are close to each other are perceived as a group.
Create a scatter plot of sepal_length vs. sepal_width and discuss how proximity groups the points.


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

Similarity:
Explain how objects that are similar are perceived as a group.
Modify the scatter plot to use different markers for each species to illustrate the principle of similarity.


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species')
plt.title('Sepal Length vs Sepal Width with Different Markers')
plt.show()
```

Continuity:
Explain how the human eye follows continuous lines and paths.
Create a line plot of average sepal_length for each species to demonstrate continuity.


```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=iris, x='species', y='sepal_length', ci=None, marker='o')
plt.title('Average Sepal Length by Species')
plt.show()
```

Closure:
Explain how the mind fills in missing information to perceive complete shapes.
Create a bar plot of the mean petal_length for each species and show how adding error bars (standard deviation) can help achieve closure.


```python
plt.figure(figsize=(8, 6))
sns.barplot(data=iris, x='species', y='petal_length', ci='sd')
plt.title('Mean Petal Length by Species with Error Bars')
plt.show()
```
