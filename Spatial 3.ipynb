```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
```

```python
# Load the breast cancer dataset
cancer_data = load_breast_cancer()

# Convert the dataset into a DataFrame
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df['target'] = cancer_data.target
```

```python
df.head()
```

```python
# Plot 1: 3D Scatter Plot of 'mean radius', 'mean texture', and 'mean smoothness'
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['mean radius'], df['mean texture'], df['mean smoothness'], 
                     c=df['target'], cmap='coolwarm', marker='o')
```

```python
# Plot 1: 3D Scatter Plot of 'mean radius', 'mean texture', and 'mean smoothness'
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['mean radius'], df['mean texture'], df['mean smoothness'], 
                     c=df['target'], cmap='coolwarm', marker='o')

# Setting labels
ax.set_xlabel('Mean Radius')
ax.set_ylabel('Mean Texture')
ax.set_zlabel('Mean Smoothness')
plt.title('3D Scatter Plot of Mean Radius, Mean Texture, and Mean Smoothness')

# Add color bar to show the mapping of color to target values
legend1 = ax.legend(*scatter.legend_elements(), title="Target")
ax.add_artist(legend1)

plt.show()
```

```python
# Plot 2: 3D Surface Plot (for illustration, we'll create a meshgrid on a smaller sample)
# Select a smaller sample for better visualization of surface plot
sample_df = df[['mean radius', 'mean texture', 'mean smoothness']].sample(100, random_state=1)
X = sample_df['mean radius']
Y = sample_df['mean texture']
Z = sample_df['mean smoothness']

# Create a meshgrid for X and Y
X, Y = np.meshgrid(np.linspace(X.min(), X.max(), 50), np.linspace(Y.min(), Y.max(), 50))
Z = np.sin(X) + np.cos(Y)  # Example function for surface plot

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Setting labels
ax.set_xlabel('Mean Radius')
ax.set_ylabel('Mean Texture')
ax.set_zlabel('Mean Smoothness (Example Function)')
plt.title('3D Surface Plot (Example Function)')

# Add a color bar to show the mapping from Z values to colors
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()
```
